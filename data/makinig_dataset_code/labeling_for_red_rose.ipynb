# RedとRoseのCSVデータからデータセットを作成するためのプログラム

```python
import pandas as pd
CSV_PATH = '../csvs/rose.csv'
df = pd.read_csv(CSV_PATH)
df
```

```python
import glob
import os 
BASE_DIR = "../raw_datas/rose/3_15start/4_6"
row_list = glob.glob(os.path.join(BASE_DIR, "*"))
row_list.sort()
print(row_list[0:3])
print(len(row_list))
```

```python
dict_list = []  
for target_dir, column in zip(row_list, df.columns):  # 列ごとに注目
    file_path_list = glob.glob(os.path.join(target_dir, "*"))
    file_path_list.sort()  # それぞれのディレクトリ内のファイルリスト
    label_list = list(df[column])*4  # ファイルのラベルリスト
    assert len(file_path_list) == len(label_list)  # ラベル数と画像数が一致するか確認

    file_label_dict = {key: value for key, value in zip(file_path_list, label_list)}
    dict_list.append({target_dir: file_label_dict})
dict_list[0]
```

```python
import shutil
OUTPUT_DIR = '../labeled_data/rose'
DOUBLE_PATH = os.path.join(OUTPUT_DIR, 'double')
SINGLE_PATH = os.path.join(OUTPUT_DIR, 'single')
# os.mkdir(OUTPUT_DIR)
# os.mkdir(DOUBLE_PATH)
# os.mkdir(SINGLE_PATH)

double_num = 0
single_num = 0
others = 0

for row_dict in dict_list:  # row_dictは列ごとのフォルダ(8つ)
    for target_dir, data_dict_list in row_dict.items():  # data_dict_listは、画像パスをkey、labelをvalueとする辞書のリスト
        for img, label in data_dict_list.items():  # imgはパス、labelはラベル
            if label == 1:
                shutil.copy(img, os.path.join(SINGLE_PATH, img.split("/")[-1]))
                single_num += 1
            elif label == 8:
                shutil.copy(img, os.path.join(DOUBLE_PATH, img.split("/")[-1]))
                double_num += 1
            else:
                print("path:{}, label:{}".format(img, label))
                others += 1
print("double images:{}".format(double_num))
print("single images:{}".format(single_num))
print("other images:{}".format(others))
```
